# Machine Learning and Computer Vision
## یادگیری ماشین چیست؟ (What is machine learning?)

یادگیری ماشین به ما امکان می‌دهد تا از داده‌ها پیش‌بینی کنیم یا از آن‌ها یاد بگیریم.

مثال‌ها:

1. استفاده از مقالات خبری امروز برای پیش‌بینی قیمت سهام فردا

2. استفاده از کلمات یک توییت برای پیش‌بینی احساس بیان‌شده (مثلاً عصبانیت، غم، شادی، دیدگاه سیاسی)

3. پیش‌بینی بیماری‌ها بر اساس داده‌های پزشکی

4. استفاده از داده‌های خرید قبلی شما برای پیش‌بینی خریدهای آینده

5. و محبوب‌ترین مثال، فیلتر کردن ایمیل‌های اسپم

## دو نوع اصلی یادگیری ماشین (Two main types of machine learning)

### یادگیری نظارت‌شده و بدون نظارت (Supervised and unsupervised learning)

یادگیری نظارت‌شده - ما به الگوریتم نمونه‌هایی با حقیقت زمینه می‌دهیم. سپس الگوریتم مدلی می‌سازد که ورودی‌ها را به خروجی‌ها نگاشت می‌کند (فرآیند آموزش). سپس از این مدل برای پیش‌بینی خروجی‌های ورودی‌های جدید استفاده می‌کنیم.

یادگیری بدون نظارت - به طور مشابه سعی می‌کند خروجی را از داده‌های ورودی پیش‌بینی کند اما هیچ نمونه با حقیقت زمینه داده نمی‌شود؛ مانند خوشه‌بندی.

## یادگیری ماشین در بینایی کامپیوتر (Machine learning in computer Vision)

چه کاربردی دارد؟

- پیش‌بینی وجود چهره در تصویر

- شناسایی چهره‌ها در تصاویر

- طبقه‌بندی نوع خودرو در یک تصویر

- شناسایی علائم جاده‌ای

- خواندن پلاک خودرو

- تأیید امضا

- خواندن اثر انگشت

- ایکس‌باکس کینکت

- شناسایی اشیاء

- درک دست‌خط

## معمولاً چگونه استفاده می‌شود؟ (How is typically used?)

- بررسی شدت پیکسل خام یا رنگ‌ها

- توصیف‌گرهای نقاط کلیدی

- هیستوگرام گرادیان‌های تصویر (HOGs)

- آمار پیکسل (مانند ابعاد، میانگین، واریانس و بسیاری دیگر)

## الگوریتم نزدیک‌ترین همسایگان (K-Nearest Neighbors Algorithm)

- KNN یک طبقه‌بند ساده یادگیری ماشین است که یک ورودی جدید را بر اساس نزدیک‌ترین نمونه‌ها در فضای ویژگی طبقه‌بندی می‌کند.

- این الگوریتم بر اساس فاصله اقلیدسی وزنی k نزدیک‌ترین همسایه‌ها عمل می‌کند.
