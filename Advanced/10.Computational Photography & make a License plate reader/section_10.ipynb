## Photo Denoising

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('images/elephant.jpg')

dst = cv2.fastNlMeansDenoisingColored(image, None, 11, 6, 7, 21)

# Convert BGR to RGB for matplotlib display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
dst_rgb = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Denoised')
plt.imshow(dst_rgb)
plt.axis('off')

plt.show()
```

در این بخش، یک تصویر رنگی با استفاده از OpenCV بارگذاری شده و سپس با استفاده از تابع `cv2.fastNlMeansDenoisingColored` نویز آن حذف شده است.  
برای نمایش صحیح تصاویر در کتابخانه matplotlib، ابتدا فضای رنگی تصاویر از BGR به RGB تبدیل شده است.  
در نهایت، هر دو تصویر اصلی و تصویر نویززدایی‌شده به صورت کنار هم نمایش داده شده‌اند تا بتوانید تفاوت کیفیت را به وضوح مشاهده کنید.  
این روش یکی از تکنیک‌های مؤثر حذف نویز غیرمحلی (Non-Local Means Denoising) برای تصاویر رنگی است که جزئیات تصویر را تا حد زیادی حفظ می‌کند.

**چهار نوع مختلف از حذف نویز Non-Local Means وجود دارد:**

- cv2.fastNlMeansDenoising() - برای تصاویر تک‌کاناله (سیاه و سفید) استفاده می‌شود.
- cv2.fastNlMeansDenoisingColored() - برای تصاویر رنگی به کار می‌رود.
- cv2.fastNlMeansDenoisingMulti() - برای دنباله‌ای از تصاویر سیاه و سفید که در مدت زمان کوتاهی ثبت شده‌اند، استفاده می‌شود.
- cv2.fastNlMeansDenoisingColoredMulti() - مشابه مورد قبل، اما برای تصاویر رنگی.
