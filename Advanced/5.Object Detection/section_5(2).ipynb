## Image Features

ویژگی‌های تصویر (Image Features) به خصوصیات یا اطلاعاتی گفته می‌شود که از یک تصویر استخراج می‌شوند و می‌توانند برای تحلیل، شناسایی یا طبقه‌بندی تصاویر مورد استفاده قرار گیرند. این ویژگی‌ها می‌توانند شامل لبه‌ها، گوشه‌ها، بافت، رنگ، شکل و سایر الگوهای موجود در تصویر باشند. استخراج ویژگی‌ها یکی از مراحل کلیدی در پردازش تصویر و بینایی ماشین است و به الگوریتم‌ها کمک می‌کند تا تصاویر را بهتر درک و تحلیل کنند.

**ویژگی‌های تصویر** نقش کلیدی در موفقیت الگوریتم‌های پردازش تصویر و بینایی ماشین ایفا می‌کنند. این ویژگی‌ها به سیستم‌ها امکان می‌دهند تا اطلاعات معنادار و قابل استفاده‌ای از تصاویر استخراج کنند و تفاوت میان اشیا، الگوها یا بافت‌های مختلف را به‌درستی تشخیص دهند.  
هرچه ویژگی‌های استخراج‌شده دقیق‌تر و مرتبط‌تر باشند، تحلیل و طبقه‌بندی تصاویر با دقت بیشتری انجام می‌شود.

### کاربردهای ویژگی‌های تصویر

- **تشخیص چهره** (Face Detection)  
- **شناسایی و ردیابی اشیا** (Object Detection and Tracking)  
- **طبقه‌بندی تصاویر** (Image Classification)  
- **جستجوی تصاویر مشابه** (Similar Image Retrieval)  
- **تحلیل ویدئو** (Video Analysis)
- و بسیاری از مسائل دیگر در حوزه پردازش تصویر و یادگیری ماشین

انتخاب و استخراج ویژگی‌های مناسب، تأثیر مستقیمی بر عملکرد و دقت مدل‌های بینایی ماشین دارد و بدون آن‌ها، این مدل‌ها قادر به ارائه نتایج قابل قبول نخواهند بود.

## What Defines an Interesting?

مناطق جالب در تصویر، بخش‌هایی هستند که اطلاعات متمایز، غنی و قابل توجهی را نسبت به سایر بخش‌ها ارائه می‌دهند. این مناطق معمولاً نقش کلیدی در تحلیل و تفسیر تصویر دارند و می‌توانند برای شناسایی، طبقه‌بندی یا ردیابی اشیا مورد استفاده قرار گیرند.

ویژگی‌های اصلی مناطق جالب در تصویر عبارت‌اند از:

- **تغییرات شدید شدت روشنایی:** نقاطی که در آن‌ها شدت روشنایی به‌طور ناگهانی تغییر می‌کند (مانند لبه‌ها و گوشه‌ها).
- **وجود ساختارهای هندسی خاص:** مانند گوشه‌ها، تقاطع خطوط یا نقاطی با بافت پیچیده.
- **الگوهای بافتی یا ساختارهای غیرتکراری:** بخش‌هایی که الگوهای منحصربه‌فرد یا بافت‌های متمایز دارند.

توجه: نویز تصویر نیز ممکن است به‌اشتباه به‌عنوان یک ویژگی جالب شناسایی شود. بنابراین، تشخیص صحیح ویژگی‌های واقعی از نویز اهمیت زیادی دارد و باید با دقت انجام شود.

انتخاب صحیح مناطق جالب، تأثیر مستقیمی بر کیفیت و دقت الگوریتم‌های پردازش تصویر خواهد داشت.

## Characteristics of Good or Interesting Image Features

ویژگی‌های خوب یا جالب تصویر باید دارای خصوصیات زیر باشند تا در کاربردهای مختلف پردازش تصویر و بینایی ماشین مؤثر واقع شوند:

- **تکرارپذیری (Repeatability):**  
    ویژگی باید در تصاویر مختلف از یک صحنه، تحت شرایط نوری یا زاویه دید متفاوت، به طور قابل اعتماد شناسایی شود.

- **تمایزپذیری (Distinctiveness):**  
    هر ویژگی باید منحصربه‌فرد باشد و به‌راحتی از سایر ویژگی‌های تصویر یا صحنه قابل تشخیص باشد تا امکان تطبیق دقیق فراهم شود.

- **کارآمدی و فشردگی (Efficiency & Compactness):**  
    تعداد ویژگی‌های استخراج‌شده باید نسبت به کل پیکسل‌های تصویر کم باشد تا پردازش سریع‌تر و ذخیره‌سازی آسان‌تر شود.

- **محلی بودن (Locality):**  
    هر ویژگی باید فقط بخش کوچکی از تصویر را پوشش دهد تا در برابر انسداد جزئی، تغییرات پس‌زمینه یا شلوغی مقاوم باشد.

- **پایداری (Robustness):**  
    ویژگی باید نسبت به نویز، تغییرات روشنایی، مقیاس، چرخش و سایر تغییرات هندسی مقاوم باشد.

این خصوصیات باعث می‌شوند ویژگی‌های انتخاب‌شده برای تحلیل، شناسایی و تطبیق تصاویر، قابل اعتماد و مؤثر باشند.

## Corners as features

گوشه‌ها زمانی شناسایی می‌شوند که با جابه‌جایی یک پنجره در هر جهتی روی آن نقطه، تغییر زیادی در شدت روشنایی مشاهده شود.

- **سطح صاف (Flat):** هیچ تغییری در هیچ جهتی وجود ندارد.
- **لبه (Edge):** تغییر فقط در یک جهت رخ می‌دهد.
- **گوشه (Corner):** تغییر در همه جهات مشاهده می‌شود.



## Finding Corners

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image then grayscale
image = cv2.imread('images/chess.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# The cornerHarris function requires the array datatype to be float32
gray = np.float32(gray)

harris_corners = cv2.cornerHarris(gray, 3, 3, 0.05)

# We use dilation of the corner points to enlarge them
kernel = np.ones((7,7),np.uint8)
harris_corners = cv2.dilate(harris_corners, kernel, iterations = 2)

# Threshold for an optimal value, it may vary depending on the image.
# Try different threshold values to find the best one for harris_corners
best_thresh = 0.0
max_corners = 0

for thresh in np.arange(0.01, 0.2, 0.01):
    corners_count = np.sum(harris_corners > thresh * harris_corners.max())
    if corners_count > max_corners:
        max_corners = corners_count
        best_thresh = thresh

# Use the best threshold found
image[harris_corners > best_thresh * harris_corners.max()] = [255, 100, 127]
print(f"Best threshold: {best_thresh}, Number of corners: {max_corners}")

# Convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,8))
plt.imshow(image_rgb)
plt.title('Harris Corners')
plt.axis('off')
plt.show()
```

این کد پیاده‌سازی الگوریتم Harris Corner Detection برای شناسایی گوشه‌ها در تصویر است. مراحل به شرح زیر است:

1. **خواندن و تبدیل تصویر به سطح خاکستری:**  
    تصویر با OpenCV خوانده شده و به تصویر خاکستری تبدیل می‌شود تا پردازش ساده‌تر شود.

2. **تبدیل نوع داده:**  
    تصویر خاکستری به نوع float32 تبدیل می‌شود، چون تابع cornerHarris به این نوع داده نیاز دارد.

3. **محاسبه گوشه‌ها با Harris:**  
    تابع `cv2.cornerHarris` برای شناسایی نقاط گوشه‌ای در تصویر استفاده می‌شود. پارامترها شامل اندازه پنجره، اندازه بلوک و مقدار k هستند.

4. **افزایش نقاط گوشه‌ای:**  
    با استفاده از عملیات dilate (گسترش)، نقاط گوشه‌ای بزرگ‌تر و واضح‌تر می‌شوند.

5. **یافتن بهترین آستانه (Threshold):**  
    با آزمون مقادیر مختلف آستانه، تعداد گوشه‌های شناسایی‌شده برای هر مقدار محاسبه می‌شود تا بهترین مقدار آستانه (که بیشترین گوشه را شناسایی می‌کند) انتخاب شود.

6. **رنگ‌آمیزی گوشه‌ها:**  
    نقاطی که مقدارشان از آستانه انتخاب‌شده بیشتر است، با رنگ خاصی (صورتی) در تصویر اصلی مشخص می‌شوند.

7. **تبدیل تصویر برای نمایش:**  
    تصویر از BGR به RGB تبدیل می‌شود تا با matplotlib به‌درستی نمایش داده شود.

8. **نمایش تصویر:**  
    تصویر نهایی با گوشه‌های مشخص‌شده نمایش داده می‌شود.

این کد به شما کمک می‌کند تا نقاط گوشه‌ای (ویژگی‌های جالب) را در یک تصویر پیدا و مشاهده کنید.


## Improved Corner Detection using - Good Features to Track

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('images/chess.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# We specific the top 50 corners
corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 150)

for corner in corners:
    x, y = corner[0]
    x = int(x)
    y = int(y)
    cv2.rectangle(img,(x-10,y-10),(x+10,y+10),(0,255,0), 2)
    
# Show with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img_rgb)
plt.title("Corners Found (matplotlib)")
plt.axis('off')
plt.show()
```

این کد پیاده‌سازی الگوریتم "Good Features to Track" برای شناسایی گوشه‌های تصویر است. توضیح مراحل به صورت زیر است:

1. **خواندن و تبدیل تصویر به خاکستری:**  
    تصویر با استفاده از OpenCV خوانده می‌شود و به تصویر خاکستری تبدیل می‌شود تا پردازش ساده‌تر شود.

2. **شناسایی بهترین گوشه‌ها:**  
    تابع `cv2.goodFeaturesToTrack` برای یافتن بهترین نقاط گوشه‌ای (در این مثال تا ۱۰۰ نقطه) با مقدار آستانه کیفیت ۰.۰۱ و حداقل فاصله ۱۵۰ پیکسل بین گوشه‌ها استفاده می‌شود. این تابع نقاطی را انتخاب می‌کند که بیشترین اطلاعات ساختاری را دارند و برای ردیابی مناسب هستند.

3. **رسم مستطیل دور گوشه‌ها:**  
    برای هر گوشه شناسایی‌شده، مختصات x و y استخراج شده و یک مستطیل سبز رنگ به اندازه ۱۰ پیکسل دور آن نقطه رسم می‌شود تا گوشه‌ها در تصویر به‌وضوح نمایش داده شوند.

4. **تبدیل تصویر برای نمایش:**  
    تصویر از BGR به RGB تبدیل می‌شود تا با matplotlib به‌درستی نمایش داده شود و رنگ‌ها به‌درستی دیده شوند.

5. **نمایش تصویر:**  
    تصویر نهایی با گوشه‌های مشخص‌شده نمایش داده می‌شود تا بتوانید نتیجه شناسایی را به صورت بصری ارزیابی کنید.

**نکات تکمیلی برای بهبود انتخاب گوشه‌ها:**

- مقدار پارامتر `maxCorners` (تعداد گوشه‌های خروجی) را می‌توانید با توجه به پیچیدگی تصویر و نیاز پروژه تنظیم کنید. مقدار بیشتر باعث شناسایی گوشه‌های بیشتر (شامل نقاط کم‌اهمیت یا نویزی) و مقدار کمتر باعث تمرکز بر مهم‌ترین گوشه‌ها می‌شود.
- مقدار `qualityLevel` (آستانه کیفیت) نیز بر حساسیت الگوریتم تأثیر می‌گذارد؛ مقدار کمتر گوشه‌های بیشتری را شناسایی می‌کند اما ممکن است نقاط ضعیف‌تر نیز انتخاب شوند.
- برای انتخاب بهترین مقادیر این پارامترها، می‌توانید چند مقدار مختلف را امتحان کنید و نتایج را به صورت بصری مقایسه کنید تا بهترین تنظیمات را برای تصویر خود بیابید.
- این روش نسبت به Harris Corner Detector معمولاً نتایج دقیق‌تر و گوشه‌های متمایزتری ارائه می‌دهد و برای کاربردهایی مانند ردیابی و تطبیق نقاط کلیدی بسیار مناسب است.

## معرفی SIFT (Scale Invariant Feature Transform)

نیاز به تحمل نسبت به تغییر مقیاس (که به آن مقیاس‌ناپذیری یا Scale Invariance گفته می‌شود) در بسیاری از کاربردهای بینایی ماشین وجود دارد.

الگوریتم SIFT یکی از پرکاربردترین روش‌ها در بینایی ماشین است که به دلیل موفقیت بالا در حل مشکل مقیاس‌ناپذیری، بسیار مورد توجه قرار گرفته است. این الگوریتم می‌تواند ویژگی‌هایی را از تصویر استخراج کند که نسبت به تغییر مقیاس، چرخش و حتی تغییرات جزئی روشنایی مقاوم هستند.

نکته مهم این است که SIFT ثبت اختراع شده بود و از نسخه ۳.۰ به بعد، به صورت رایگان در OpenCV در دسترس نیست.

## SIFT به طور خلاصه

1. ابتدا نقاط کلیدی جالب را در تصویر با استفاده از روش تفاوت گاوسی (Difference of Gaussian) شناسایی می‌کنیم. این نواحی از تصویر جایی هستند که تغییرات شدت روشنایی از یک آستانه خاص بیشتر بوده و نسبت به لبه‌ها ویژگی‌های متمایزتری دارند.

2. سپس برای این نواحی جالب، یک توصیف‌گر برداری (Vector Descriptor) ایجاد می‌کنیم. مقاومت نسبت به تغییر مقیاس (Scale Invariance) از طریق فرآیند زیر حاصل می‌شود:

- نقاط کلیدی در چندین مقیاس مختلف بررسی می‌شوند.

-  مقیاسی که در آن معیار پایداری خاصی برآورده می‌شود، انتخاب شده و در توصیف‌گر برداری ثبت می‌شود. بنابراین، صرف‌نظر از اندازه اولیه، پایدارترین مقیاس انتخاب می‌شود که باعث مقاومت نسبت به تغییر مقیاس می‌گردد.

3. مقاومت نسبت به چرخش (Rotation Invariance) با تعیین جهت‌گیری (Orientation) نقطه کلیدی با استفاده از گرادیان شدت تصویر به دست می‌آید. زمانی که جهت دوبعدی را بدانیم، می‌توانیم این جهت را نرمال‌سازی کنیم.
