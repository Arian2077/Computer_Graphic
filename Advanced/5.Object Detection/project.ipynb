# Finding Barny

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r'images\crowd.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# load Template image
template = cv2.imread(r'images\Barny.png')
template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)

# Check template size is not larger than source image
if template_gray.shape[0] > gray.shape[0] or template_gray.shape[1] > gray.shape[1]:
    raise ValueError("Template image is larger than source image.")

result = cv2.matchTemplate(gray, template_gray, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

# Create Bounding Box
top_left = max_loc
bottom_right = (top_left[0] + 80, top_left[1] + 80)
cv2.rectangle(image, top_left, bottom_right, (0,0,255), 5)

# Show images using matplotlib subplots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(cv2.imread(r'images\crowd.jpg'), cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Detected Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

این کد با استفاده از کتابخانه‌های OpenCV و Matplotlib، مراحل زیر را انجام می‌دهد:

1. **بارگذاری تصاویر:** تصویر اصلی و تصویر الگو (Barny) بارگذاری و به حالت خاکستری تبدیل می‌شوند تا پردازش ساده‌تر و سریع‌تر انجام شود.
2. **بررسی ابعاد الگو:** اطمینان حاصل می‌شود که ابعاد تصویر الگو از تصویر اصلی بزرگ‌تر نباشد.
3. **یافتن محل بیشترین شباهت:** با استفاده از تابع `matchTemplate`، محل بیشترین شباهت بین تصویر اصلی و الگو شناسایی می‌شود و مختصات آن به دست می‌آید.
4. **رسم کادر دور ناحیه شناسایی‌شده:** با توجه به مختصات به‌دست‌آمده (`top_left` و `bottom_right`)، یک کادر قرمز رنگ دور ناحیه‌ای که Barny در تصویر اصلی قرار دارد رسم می‌شود.
5. **نمایش نتایج:** با استفاده از Matplotlib، تصویر اصلی و تصویر با کادر مشخص‌شده به صورت کنار هم نمایش داده می‌شوند تا محل دقیق Barny به وضوح قابل مشاهده باشد.
