# introduction

یکی از جذاب‌ترین جنبه‌های برنامه‌نویسی، توانایی بخشیدن به کامپیوتر برای دیدن و تشخیص تصاویر است. امروزه با پیشرفت تکنولوژی، این امکان به‌سادگی فراهم شده است. با این حال، تشخیص تصاویر توسط ماشین‌ها با چالش‌های متعددی همراه است، زیرا تصاویر می‌توانند مشکلات مختلفی داشته باشند که فرآیند تشخیص را دشوار می‌کنند.

برخی از این مشکلات عبارتند از:

1. محدودیت‌های لنز و سنسور دوربین‌ها
2. زاویه دید نامناسب تصاویر
3. تغییرات در شدت نور (کم‌نور یا پرنور بودن)
4. تغییرات در اندازه تصاویر
5. تغییر شکل‌های غیرطبیعی
6. انسداد بخشی از تصویر
7. شلوغی و پیچیدگی پس‌زمینه
8. وجود مدل‌ها و نسخه‌های مختلف از یک شیء یا موجودیت
9. تصاویر مبهم و توهمات نوری

با وجود این چالش‌ها، استفاده از تکنیک‌ها و الگوریتم‌های پیشرفته می‌تواند به مدیریت و حل این مشکلات کمک کند. این چالش‌ها نه تنها موانع هستند، بلکه فرصتی برای توسعه راه‌حل‌های خلاقانه و بهبود توانایی‌های ماشین‌ها در تشخیص تصاویر محسوب می‌شوند.

## Why opencv in python

1. پایتون یکی از ساده‌ترین و محبوب‌ترین زبان‌های برنامه‌نویسی است که یادگیری و استفاده از آن بسیار آسان است.
2. این زبان در حوزه‌های علم داده و یادگیری ماشین عملکرد بسیار قدرتمندی دارد و ابزارهای متنوعی برای این کاربردها ارائه می‌دهد.
3. تصاویر در پایتون با استفاده از کتابخانه numpy به صورت آرایه ذخیره می‌شوند، که این ویژگی امکان انجام عملیات پیچیده بر روی تصاویر را به‌سادگی فراهم می‌کند.

## Setup

1. ابتدا Python را از سایت رسمی [python.org](https://www.python.org) دانلود و نصب کنید.
2. برای راحتی بیشتر، می‌توانید [Anaconda](https://www.anaconda.com) را دانلود و نصب کنید. در غیر این صورت، می‌توانید از محیط‌های دیگری مانند Visual Studio Code (VSC) نیز استفاده کنید.
3. کتابخانه OpenCV را با استفاده از دستور زیر نصب کنید:
    ```
    pip install opencv-contrib-python
    ```
4. اکنون می‌توانید کار با این کتابخانه قدرتمند را آغاز کنید!
