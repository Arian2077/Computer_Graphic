# Computer Graphic Session 2

```python
from matplotlib import pyplot as plt
import cv2
```

## Using blurred filter

```python
# Reading the image
image = cv2.imread("image.jpg")
RGB = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
# Using effect
blurred = cv2.GaussianBlur(image,(9,9),10)
blurred_rgb=cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB)
# Showing the images in Jupyter using matplotlib
# Showing The Original image
plt.imshow(RGB)
plt.title("Original")
plt.axis("off")
plt.show()
# Showing The Blured image
plt.imshow(blurred_rgb)
plt.title("Image with Gaussian Blur")
plt.axis("off")
plt.show()
```

## Showing Gray & RGB colored pictures besides each other

```python
# Converting BGR to Gray color
Gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
# Setting the figure size
plt.figure(figsize=(12,6)) 
# Using subplot
plt.subplot(1,2,1) # 1 row / 2 columns
plt.imshow(RGB)
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(Gray,cmap='gray')
plt.title("Gray")
plt.axis('off')
plt.show()
```

## Comparison between Gaussian & Average blur 

```python
# Applying Average Blur
avg_blur = cv2.blur(image, (9, 9))
avg_blur = cv2.cvtColor(avg_blur, cv2.COLOR_BGR2RGB)

# Setting the figure size
plt.figure(figsize=(12,6)) 

# Using subplot
plt.subplot(1, 2, 1)  # 1 row / 2 columns / 1 place
plt.imshow(avg_blur)
plt.title("Average Blur")
plt.axis('off')

plt.subplot(1, 2, 2) # 1 row / 2 columns / 2 place
plt.imshow(blurred_rgb)
plt.title("Gaussian Blur")
plt.axis('off')

plt.show()
```

## Adding noise then using median filter

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Assuming 'image' is already defined and is a valid image
noisy = image.copy()
rows, cols, _ = noisy.shape

# Add noise
for i in range(80000):
    x = np.random.randint(0, rows)  # Row index
    y = np.random.randint(0, cols)   # Column index
    noisy[x, y] = [255, 255, 255] if i % 2 == 0 else [0, 0, 0]

# Using median blur
median = cv2.medianBlur(noisy, 5)
# Setting the figure size
plt.figure(figsize=(12,6)) 
# Using subplot
plt.subplot(1, 2, 1)  # 1 row / 2 columns / 1 place
plt.imshow(cv2.cvtColor(noisy, cv2.COLOR_BGR2RGB))
plt.title("Noisy Image")
plt.axis('off')

plt.subplot(1, 2, 2)  # 1 row / 2 columns / 2 place
plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
plt.title("Median Blur")
plt.axis('off')

plt.show()
```
