```python
from matplotlib import pyplot as plt
import cv2
```

## Using blurred filter

```python
# Reading the image
image = cv2.imread("image.jpg")
RGB = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
# Using effect
blurred = cv2.GaussianBlur(image,(9,9),20)
blurred_rgb=cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB)
# Showing the images in Jupyter using matplotlib
plt.imshow(RGB)
plt.title("Original")
plt.axis("off")
plt.show()

plt.imshow(blurred_rgb)
plt.title("Image with Gaussian Blur")
plt.axis("off")
plt.show()
```
