# Computer graphic

```python
from matplotlib import pyplot as plt
```

## Making a simple Liner vector

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.title("Simple Line Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

## read and show a image

```python
import cv2

# Reading the image
image = cv2.imread('nature.jpg')

# Converting image color from BGR to RGB
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

# Showing image in Jupyter
plt.imshow(image)
plt.axis("off")
plt.title("Image display")
plt.show()
```
