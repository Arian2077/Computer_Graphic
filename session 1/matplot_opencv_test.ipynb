# Computer graphic

```python
from matplotlib import pyplot as plt
```

## Making a simple Liner vector

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.title("Simple Line Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

## read and show a image

```python
import cv2

# Reading the image
image = cv2.imread('nature.jpg')

# Converting image color from BGR to RGB
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

# Showing image in Jupyter
plt.imshow(image)
plt.axis("off")
plt.title("Image display")
plt.show()
```

## Converting image to gray color

```python
# Doing this after you read the image with cv2
# Turning color from BGR to GRAY
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# Showing the image
plt.imshow(gray,cmap='gray')
plt.title("Gray scale image")
plt.axis('off')
plt.show()
```

## Resizing the image

```python
# Doing this after you read the image with cv2 and convert the color from BGR to RGB
# Changing the size of image from 1920*1080 to 640*640
resized = cv2.resize(image,(640,640))

# Showing the resized image
plt.imshow(resized)
plt.title("Resized image")
plt.axis("off")
plt.show()
```
